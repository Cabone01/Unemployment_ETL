# US Unemployed Data

```python
us_df = pd.read_csv("../Datasets/aat1_csv.csv")
us_df.head()
```

```python
us_df = us_df.drop(columns=["footnotes"])
us_df.head()
```

```python
us_df.dtypes
```

```python
try:
    cnx = mysql.connector.connect(host='localhost', database='unemployment_usa', user='taco', password=key)
    if cnx.is_connected():
        cursor = cnx.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        print("You're connected to database: ", record)
        
        cursor.execute('DROP TABLE IF EXISTS unemployment_usa_data;')
        print('Creating table....')

        cursor.execute("CREATE TABLE unemployment_usa_data(yr YEAR,population INT,labor_force INT,population_prct FLOAT,employed_total INT,employed_prct FLOAT,agriculture_ratio INT,nonagriculture_ratio INT,unemployed INT,unemployed_prct FLOAT,not_in_labor INT);")
        print("Table is created....")

        for i,row in us_df.iterrows():
            sql = "INSERT INTO unemployment_usa.unemployment_usa_data VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            print("Record inserted")
            cnx.commit()
except Error as e:
            print("Error while connecting to MySQL", e)
```

```python
# Execute query
sql = "SELECT * FROM unemployment_usa.unemployment_usa_data"
cursor.execute(sql)
# Fetch all the records
result = cursor.fetchall()
for i in result:
    print(i)
```

# US Demographic and Data merge

```python
us_df_mrge = us_df[["year", "population", "employed_total", "unemployed"]]
```

```python
us_demo_unemp_df = us_df_mrge.merge(us_demo_df, left_on="year", right_on="Year", how="inner")
us_demo_unemp_df
```

```python
us_demo_unemp_df = us_demo_unemp_df.drop(columns=["Year"])
us_demo_unemp_df.head()
```

```python
column = us_demo_unemp_df.pop("Month")
us_demo_unemp_df.insert(1, "month", column)
```

```python
try:
    cnx = mysql.connector.connect(host='localhost', database='unemployment_usa', user='taco', password=key)
    if cnx.is_connected():
        cursor = cnx.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        print("You're connected to database: ", record)
        
        cursor.execute('DROP TABLE IF EXISTS unemployment_demographic_prct_2010;')
        print('Creating table....')
        
        cursor.execute("CREATE TABLE unemployment_demographic_prct_2010(year YEAR, month VARCHAR(12), population INT, employed_total INT, unemployed_total INT, primary_school FLOAT, high_school FLOAT, associates_degree FLOAT, professional_degree FLOAT, white FLOAT, black FLOAT, asian FLOAT, hispanic FLOAT, men Float, women FLOAT);")
        print("Table is created....")
        
        for i,row in us_demo_unemp_df.iterrows():
            sql = "INSERT INTO unemployment_usa.unemployment_demographic_prct_2010 VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            print("Record inserted")
            cnx.commit()
except Error as e:
            print("Error while connecting to MySQL", e)
```

```python
# Execute query
sql = "SELECT * FROM unemployment_usa.unemployment_demographic_prct_2010"
cursor.execute(sql)
# Fetch all the records
result = cursor.fetchall()
for i in result:
    print(i)
```
